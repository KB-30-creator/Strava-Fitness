```python
import matplotlib.pyplot as plt # Convert the ActivityDate to datetime
```

```python
hourly_data['ActivityHour'] = pd.to_datetime(hourly_data['ActivityHour'])
```

```python
# Count unique users per day
users_per_day = hourly_data.groupby(hourly_data['ActivityHour'].dt.date)['Id'].nunique().reset_index()
users_per_day.columns = ['Date', 'UserCount']
```

```python
# Create ActiveHours flag
hourly_data['ActiveFlag'] = hourly_data['StepTotal'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Calculate Active Hours per Day
active_hours_per_day = hourly_data.groupby(hourly_data['ActivityHour'].dt.date)['ActiveFlag'].sum().reset_index()
active_hours_per_day.columns = ['Date', 'ActiveHours']
```

```python
# Merge with active hours
active_hours_per_day = active_hours_per_day.merge(users_per_day, on='Date')
active_hours_per_day['AvgActiveHoursPerUser'] = active_hours_per_day['ActiveHours'] / active_hours_per_day['UserCount']
```

```python
print(active_hours_per_day.head())
```

```python
# Plot
plt.figure(figsize=(12, 5))
plt.plot(active_hours_per_day['Date'], active_hours_per_day['AvgActiveHoursPerUser'], color='blue', marker='o')
plt.xlabel('Date')
plt.ylabel('Avg Active Hours Per User')
plt.title('Average Active Hours Per User Per Day')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(daily_data['ActivityDate'], daily_data['DailyCalories'], color='red', label='DailyCalories', marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Calories')
plt.title('Calories Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# % of Users Averaging 10,000+ Steps
user_avg_steps = daily_data.groupby("Id")["TotalSteps"].mean()
highly_active = (user_avg_steps >= 10000).sum()
total_users = user_avg_steps.count()
```

```python
active_percent = round(highly_active / total_users * 100,2)
print(f"{active_percent}% of users are highly active(10000+ steps/day)")
```

```python
# % of Total Steps by Month
daily_data['Month'] = pd.to_datetime(daily_data['ActivityDate']).dt.month
monthly_steps = daily_data.groupby("Month")["TotalSteps"].sum()
monthly_percent = round(monthly_steps / monthly_steps.sum() * 100, 2)
monthly_percent.plot(kind='bar', color='pink', title='% of Total Steps by Month')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```
